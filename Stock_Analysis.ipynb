```python
import numpy as np
import pandas as pd
from pandas_datareader import data, wb
import datetime
%matplotlib inline
```

```python
bank_stocks = pd.read_pickle('all_banks') 
```

```python
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
returns = pd.DataFrame()
```

```python
tickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
for tick in tickers:
    returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
returns.head()
```

```python
import seaborn as sns
sns.pairplot(returns[1:])
```

```python
# Worst_Drop
returns.idxmin()
```

```python
# Best Single Day Gain
# Citigroup stock split in May 2011 and JPM day after inauguration.
returns.idxmax()
```

```python
returns.std() # Citigroup_Riskiest
```

```python
returns.ix['2015-01-01':'2015-12-31'].std() # Very similar risk profiles, but Morgan Stanley or BofA
```

```python
sns.distplot(returns.ix['2015-01-01':'2015-12-31']['MS Return'],color='green',bins=100)
```

```python
sns.distplot(returns.ix['2008-01-01':'2008-12-31']['C Return'],color='red',bins=100)
```
